```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Ponemos valor a las variables involucradas
m = 1
k = 1

# Establecemos una malla de valores de -10 a 10 con saltos de 1 en 1 para cada una de las variables
x,v = np.meshgrid(np.arange(-10,10,1),np.arange(-10,10,1))

# Definimos las derivadas de cada variable
x_dot = v
v_dot = -k/m * x

# Graficamos el campo vectorial

# plt.quiver funciona así: 
#      plt.quiver( %ejex, %ejey, %dirección_flecha_x, %dirección_flecha_y, %parametros_adicionales )

plt.quiver(x,v,x_dot,v_dot,units='xy',scale=10,color='red')

# Configuración del Gráfico

## Titulo del gráfico
plt.title('Campo vectorial asociado',fontsize=16)

## Nombre de los ejes
plt.xlabel('x')
plt.ylabel('v')

## Limite de los ejes
plt.xlim(-10,10)
plt.ylim(-10,10)

## Tamaño de la figura
plt.gcf().set_size_inches(7,7)

## Lineas centrales
plt.axvline(x=0, color="black", linestyle='dashed', linewidth=0.2)
plt.axhline(y=0, color="black", linestyle='dashed', linewidth=0.2)

# Mostrar Gráfico
plt.show()
```
