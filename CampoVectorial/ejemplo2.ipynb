# Curvas del espacio fase

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 1000)
x_punto = np.linspace(-5, 5, 1000)

X, X_dot = np.meshgrid(x,x_punto)

E = X**3 + 1/X + X_dot**2 # Ecuación de energía

plt.contour(X,X_dot,E,levels = 10)

plt.show()
```
